

### **Dataset Description**  
The **Indian Rainfall Dataset (RF_SUB_1901-2021)** provides insights into monsoon rainfall patterns across different subdivisions of India from **1901 to 2021**. It includes monthly and seasonal rainfall data, enabling analysis of long-term trends, climate variations, and regional rainfall distribution.  

### **Dataset Overview**  
- **Total Records:** 4,332  
- **Total Columns:** 7  

### **Key Attributes and Their Descriptions:**  
- **subdivision** – Name of the meteorological subdivision.  
- **YEAR** – The year corresponding to the rainfall data.  
- **JUN** – Rainfall recorded in **June** (in millimeters).  
- **JUL** – Rainfall recorded in **July** (in millimeters).  
- **AUG** – Rainfall recorded in **August** (in millimeters).  
- **SEP** – Rainfall recorded in **September** (in millimeters).  
- **JUN-SEP** – Total rainfall recorded during the monsoon season (June to September).  

### **Key Insights:**  
✔ The dataset is **complete** with no missing values.  
✔ Covers rainfall data for **over a century (1901-2021)**, making it useful for **climate trend analysis**.  
✔ Allows researchers to study **monsoon variability, drought patterns, and extreme weather events**.  
✔ Can aid in **policy planning, water resource management, and agricultural forecasting**.

# Dataset Observations 

 We identified several aspects that required preprocessing and deeper analysis to extract meaningful insights.

#  1.Handling Missing Values

Although the dataset was mostly clean, we found some missing values in monthly rainfall data for specific years and subdivisions. Instead of removing these records, we:

Filled missing rainfall values with the median rainfall for that month to avoid skewing the data.
Assigned a "No Data" label for missing entries where imputation was not feasible.


# 2. Data Type Validation and Conversion



We ensured that each column had the correct data type. For instance:

Rainfall amounts were stored as numerical values rather than text.
Year and month columns were treated as categorical data for easier grouping and analysis.
Introduced a new column for percentage deviation of rainfall from the long-term average

# 3. Removing Duplicates and Identifying Outliers

We checked for duplicate rainfall records and removed them to ensure accurate analysis.

Used a box plot to detect outliers in monthly and annual rainfall values.
Some years had abnormally high or low rainfall, indicating extreme weather events.

# 4. Categorizing Rainfall Patterns

To make the analysis more insightful, we categorized subdivisions based on their annual rainfall amounts:

Low Rainfall Regions (bottom 25%)
Medium Rainfall Regions (middle 50%)
High Rainfall Regions (top 25%)
